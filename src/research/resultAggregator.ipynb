### Load the rankings

```python
import json

pagerank_json = open('pagerank_results.json')
elo_json = open('elo_results.json')
colley_json = open('colley_results.json')

pagerank = json.load(pagerank_json)
elo = json.load(elo_json)
colley = json.load(colley_json)

pagerank_json.close()
elo_json.close()
colley_json.close()
```

### Merge rankings

```python
from pprint import pprint

rankings = []

pagerank_predictability = pagerank['predictability']
elo_predictability = elo['predictability']
colley_predictability = colley['predictability']

predictability_constant = pagerank_predictability + elo_predictability + colley_predictability

for i in range(len(pagerank['rankings'])):
    pageranking = pagerank['rankings'][i]
    elo_ranking = next((element for element in elo['rankings'] if element['player'] == pageranking['player']), None)
    colley_ranking = next((element for element in colley['rankings'] if element['player'] == pageranking['player']), None)

    if elo_ranking == None or colley_ranking == None:
        raise Exception(f"Index {i} has does not have all 3 rankings")

    rankings.append({
        'player': pageranking['player'],
        'rating': (pageranking['ranking'] * (pagerank_predictability/predictability_constant)) + (elo_ranking['ranking'] * (elo_predictability/predictability_constant)) + (colley_ranking['ranking'] * (colley_predictability/predictability_constant))
    })
```

### Sort and print

```python
sorted_ranks = sorted(rankings, key=lambda player: player['rating'])

print('\n\n************** Aggregate **************\n')
print('===========================')
print('Rank   Rating    Player   ')
print('===========================')

for i in range(50):
    print(f'{i+1:4d}   {sorted_ranks[i]["rating"]:.5f}   {sorted_ranks[i]["player"]}')

print('')   # extra carriage return
```
