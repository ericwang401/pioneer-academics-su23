```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('./data/atp_all_matches.csv')

# Specify the name you want to search for
target_name = 'Novak Djokovic'  # Replace with the name you're interested in

# Filter the DataFrame to get rows where the specified name is either the winner or loser
filtered_df = df[(df['winner_name'] == target_name) | (df['loser_name'] == target_name)]

# Sort the filtered DataFrame by 'tourney_id' in descending order to get the games from recent to oldest
filtered_df = filtered_df.sort_values(by='tourney_id', ascending=False)

tourney_ids = filtered_df['tourney_id'].tolist()
# Display the result
print(tourney_ids)
```
